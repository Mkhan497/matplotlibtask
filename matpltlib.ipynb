```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Huzaifa Plot')
plt.plot(x, y,color='red', linestyle='--', marker='o')
plt.grid(True)
plt.legend(['Data Points'])
plt.show()
```

```python
# Create and display a histogram plot
import numpy as np
data = np.random.randn(100)  # Generate 100 random numbers from a normal distribution
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
# Plot a bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]
plt.barh(categories, values, color='red')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart Example')
plt.show()
```

PIE CHART

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels)
plt.title('Fruit Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)          # Correct: subplot with 1 row, 2 columns, subplot 1
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Subplot 1')

plt.subplot(1, 2, 2)          # Correct: 1 row, 2 columns, subplot 2
plt.bar([1, 2, 3], [5, 7, 3])
plt.title('Subplot 2')

plt.tight_layout()            # Adjust layout to prevent overlap
plt.show()
```

```python
plt.subplot(2, 2, 1)          # 2 rows, 2 columns, first subplot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Subplot 1')
plt.subplot(2, 2, 2)          # 2 rows, 2 columns, second subplot
plt.bar([1, 2, 3], [5, 7, 3])
plt.title('Subplot 2')          
plt.subplot(2, 2, 3)          # 2 rows, 2 columns, third subplot
plt.scatter([1, 2, 3], [6, 5, 4 ])
plt.title('Subplot 3')
plt.subplot(2, 2, 4)          # 2 rows, 2 columns, fourth subplot
plt.plot([1, 2, 3], [3, 2, 1])
plt.title('Subplot 4')
plt.tight_layout()            # Adjust layout to prevent overlap
```

```python
# Box plot example
import numpy as np
data = np.random.normal(0, 1, 100)
plt.boxplot(data)
plt.title('Box Plot Example')
plt.ylabel('Value')
plt.show()
```

```python
# Horizontal box plot example
import numpy as np
data = np.random.normal(0, 1, 100)
plt.boxplot(data, vert=False)
plt.title('Horizontal Box Plot Example')
plt.xlabel('Value')
plt.show()
```
